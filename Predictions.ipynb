```python
from pathlib import Path
import pandas as pd
import csv
from google.colab import files
from google.colab import drive
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
#Upload the dataset
uploaded = files.upload()
```

```python
pred_dm_df = pd.read_csv('final_df_rn.csv')
pred_dm_df.head()
```

```python
#scale sleep_hours, age, gender, gender2
scaled = StandardScaler().fit_transform(pred_dm_df[['sleep_hours', 'age', 'gender', 'gender2']])
scaled_df = pd.DataFrame(scaled, columns=['sleep_hours', 'age', 'gender', 'gender2'])
scaled_df.head()
```

```python
#concatinate scaled_df with pred_dm_df replacing the pred_dm_df values with the scaled values
pred_dm_df[['sleep_hours', 'age', 'gender', 'gender2']] = scaled_df
pred_dm_df.head()
```

```python
#separate the features (x) from the target (Y)
y = pred_dm_df["diabetes"]
X = pred_dm_df.drop(columns="diabetes")
```

```python
#create the training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_train.shape
```

```python
logistic_regression_model = LogisticRegression(max_iter = 2000, random_state = 9, solver = 'liblinear')
lr_model = logistic_regression_model.fit(X_train, y_train)
```

```python
#generate the training predictions
training_predictions = lr_model.predict(X_train)

#generate the testing predictions
testing_predictions = logistic_regression_model.predict(X_test)
```

```python
#create and save the confusion matrix for the training data
training_matrix = confusion_matrix(y_train, training_predictions)
```

```python
#create and save the confusion matrix for the testing data
testing_matrix = confusion_matrix(y_test, testing_predictions)
```

```python
#create and save the training classification report
training_report = classification_report(y_train, training_predictions)
print(training_report)
```

```python
#create and save the testing classification reprot
testing_report = classification_report(y_test, testing_predictions)
print(testing_report)
```

```python
#get the feature importance array
importances = np.abs(logistic_regression_model.coef_[0])
importance_df = pd.DataFrame({'Feature': X.columns, 'Importance':importances})
importance_df_sorted = importance_df.sort_values('Importance', ascending=False)
top_10_importances = importance_df_sorted.head(10)
print(top_10_importances)


```
